### Exercise 1: Line Plot Customization

```python
# TASK: Create a line plot with x values ranging from 0 to 10 and y values as the square of x.
# Customize the plot by adding a title, labels for both axes, and a grid.
```

```python

import matplotlib.pyplot as plt
import numpy as np

# Create an array of x values from 0 to 10
x = np.arange(0, 11)

# Calculate the corresponding y values as the square of x
y = x ** 2

# Create the line plot
plt.plot(x, y)

# Customize the plot
plt.title('Square of x vs x')  # Add title
plt.xlabel('x')  # Add x-axis label
plt.ylabel('x^2')  # Add y-axis label
plt.grid(True)  # Add grid

# Display the plot
plt.show()
```
